```python
from model import PhonemeRecogniser

def train_model(params, compute_edit_distances=False):
    p = PhonemeRecogniser(params['data_dir'])
    x_train, x_val, x_test, y_train, y_val, y_test = p.load_data(params)

    p.build_model(params)
    history = p.train_model(x_train, y_train, x_val, y_val)
    p.evaluate(x_test, y_test, history, compute_edit_distances=compute_edit_distances)
```

```python
params = {'lr': 0.0001, 'n_epochs': 2, 'batch_size': 256, 'activation': 'relu', 'dropout_rate': 0, \
          'n_hidden_layers': 3, 'n_hidden_nodes': [256, 256, 256], 'optimiser': 'adam', \
          'normalisation': 0, 'features': 'lmfcc', 'dynamic': False, 'model_name': 'test', \
          'data_dir': 'data'}

train_model(params)
```
